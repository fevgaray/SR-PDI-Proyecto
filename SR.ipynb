Visualizacion simple de las imágenes en el notebook, donde se pueden apreciar el aumento de resolucion segun lo indicado por la configuración:

```python
def mostrar_imagenes_segun_lista(ruta_carpeta, lista_nombres_archivos, EscalaResolution):
    # Obtener la lista de archivos en la carpeta
    archivos = os.listdir(ruta_carpeta)

    # Filtrar solo los archivos de imagen cuyos nombres están en la lista
    archivos_imagen = [archivo for archivo in archivos if archivo in lista_nombres_archivos]

    # Mostrar cada imagen
    for archivo_imagen in archivos_imagen:
        ruta_completa = os.path.join(ruta_carpeta, archivo_imagen)
        archivo_result = ruta_completa[:-4] + '_result' + ruta_completa[-4:]
        # Abrir la imagen con PIL
        imagen1 = Image.open(ruta_completa)
        imagen2 = Image.open(archivo_result)
        # Mostrar la imagen con matplotlib
        f, axarr = plt.subplots(1,2)
        axarr[0].imshow(imagen1)
        axarr[0].set_title("Original")
        axarr[1].imshow(imagen2)
        axarr[1].set_title("Resultado")
        plt.show()


#Ruta de la carpeta que contiene las imágenes
if EscalaResolution == 2:
    ruta_carpeta_imagenes = "dcscn-super-resolution-master/output/dcscn_L12_F196to48_NIN_A64_PS_R1F32"
if EscalaResolution != 2:
    ruta_carpeta_imagenes = "dcscn-super-resolution-master/output/dcscn_L12_F196to48_Sc"+str(EscalaResolution)+"_NIN_A64_PS_R1F32"

    
#Lista de nombres de archivos que deseas mostrar
lista_nombres_archivos = inputImagenes

#Llamar a la función para mostrar las imágenes según la lista de nombres
mostrar_imagenes_segun_lista(ruta_carpeta_imagenes, lista_nombres_archivos, EscalaResolution)
```

## Seccion 2: Herramientas Validación

Calculo de PSNR para utilizar de metrica para validar resultados:
Se crea un dataframe para una mejor visualizacion de estos calculos y poder trabajarlos mas a futuro.

```python
import pandas as pd
from skimage.metrics import peak_signal_noise_ratio


def add_to_pandas(df, image_real, image_test, image_num, scale):
    psnr = peak_signal_noise_ratio(image_real, image_test)
    print(psnr, image_num, scale)
    new_row = {'Image #': image_num, 'Scale': scale, 'PSNR':psnr}
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    return df


def calcular_PSNR_All(lista_nombres_archivos):

    df = pd.DataFrame()
    df['Image #'] = []
    df['Scale'] = []
    df['PSNR'] = []

    #Obtener la lista de carpetas en los outputs de las 3 escalas 
    carpetasOutput = os.listdir('dcscn-super-resolution-master/output')

    #Obtener images true HR que despues se van a analizar con PSNR en escalas x2 x3 x4
    imagenesHR = os.listdir('ImagesHR')

    
    for carpeta in carpetasOutput:
        # Direccion que se utilizara para obtener direccion de cada imagen resultante por cada escala
        archivos = os.listdir('dcscn-super-resolution-master/output/'+ carpeta)

        # Filtrar solo los archivos de imagen cuyos nombres están en la lista
        archivos_imagen = [archivo for archivo in archivos if archivo in lista_nombres_archivos]
        print(archivos_imagen)
        
        for imagen in archivos_imagen:
            ruta_completa = os.path.join('dcscn-super-resolution-master/output',carpeta, imagen)
            #print(ruta_completa)
            archivo_result = ruta_completa[:-4] + '_result' + ruta_completa[-4:]

            #Abrir cada imagen para calcular PSNR
            imgHR = Image.open('ImagesHR/'+ imagen)
            a = np.asarray(imgHR)
            imgSR = Image.open(archivo_result)
            b = np.asarray(imgSR)

            imgNum = lista_nombres_archivos.index(imagen)+1
            scale = carpetasOutput.index(carpeta)+2
            
            df = add_to_pandas(df, a, b, imgNum, scale)
            #print(peak_signal_noise_ratio(a,b))

    return df

theDF = calcular_PSNR_All(inputImagenes)
```

```python
#Tabla ordenada por imagenes
sorted_df = theDF.sort_values(by='Image #')
sorted_df.to_csv('output.csv', index=False)
sorted_df
```

PSNR Promedio segun resultados por escala:

```python
#Codigo para obtener PSNR promedio de cada escala
df_avg = theDF
result_df = df_avg.groupby('Scale')['PSNR'].mean().reset_index()

print(result_df)
```
